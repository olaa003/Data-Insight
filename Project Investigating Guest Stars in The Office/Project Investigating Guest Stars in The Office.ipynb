### Exploratory data analysis to see how popularity of the show varied with time

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
_=plt.plot(office['Date'],office['Viewership'])
_=plt.title('Viewership Changes over Time')
_=plt.ylabel('Viewership In Millions')
plt.show()
```

```python
#using median as a summary statistic and grouping over the seasons.
#season 5 is in 2009 and the decline starts from there.
view_average = office.groupby('Season')['Viewership'].median()
_=plt.bar(view_average.index,view_average)
```
