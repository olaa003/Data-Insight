```python
import numpy as np
np.random.seed(3)
population=np.arange(0,100) #creating a population of 100
sample= np.random.choice(population, 5) #selecting a random sample of 5 from the population

print(f'polulation is \n {population} \n\n and sample is \n {sample}')
```

```python
import matplotlib.pyplot as plt

np.random.seed(3)
normal= np.random.normal(20,1,size=10000)
poisson = np.random.poisson(2,size=10000)
exponential = np.random.exponential(5,size=10000)
binomial = np.random.binomial(20,0.3,size=10000)
geometric = np.random.geometric(0.7,size=10000)

fig,ax = plt.subplots(5)
fig.set_figheight(15)
fig.set_figwidth(15)
ax[0].hist(normal, bins=20,density=True)
ax[0].set_title('Normal distribution')
ax[1].hist(poisson, bins=20,density=True)
ax[1].set_title('Poisson distribution')
ax[2].hist(exponential, bins=20,density=True)
ax[2].set_title('Exponantial distribution')
ax[3].hist(binomial,bins=20,density=True)
ax[3].set_title('Binomial distribution')
ax[4].hist(geometric,density=True)
ax[4].set_title('Geometric distribution')
plt.show()
```

```python
# central limit theorem
np.random.seed(3)
poisson = np.random.poisson(10,size=100)
plt.hist(poisson)
plt.show()
sample_mean=[]
for i in range(100):
    sample_mean.append(np.mean(np.random.choice(poisson,100)))
plt.hist(sample_mean)
plt.show()
```
